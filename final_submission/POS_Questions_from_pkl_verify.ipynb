```python
# Test reading questions and POS tags from pkl files, run stats on cities to see subject domain
#
# Tested on Python 3.5.4
#
import numpy as np     # support for large data structures
import pandas as pd    # data structure support
import time
import collections
import re
from collections import defaultdict, Counter
```

```python
def pretty_timedelta(fmt="%d:%02d:%02d", since=None, until=None):
    """Pretty-print a timedelta, using the given format string."""
    since = since or time.time()
    until = until or time.time()
    delta_s = until - since
    hours, remainder = divmod(delta_s, 3600)
    minutes, seconds = divmod(remainder, 60)
    return fmt % (hours, minutes, seconds)
```

```python
f = "Data/train_lite0.df.pkl.gz"
train_lite = pd.read_pickle(f)
```

```python
t0 = time.time()
i = 1000
while True:
    f = "Data/train_lite"+str(i)+".df.pkl.gz"
    try:
        x = pd.read_pickle(f)
        train_lite = pd.concat([train_lite, x])
        print("Finished ", f, i, i+1000, pretty_timedelta(since=t0))
        i += 1000
    except FileNotFoundError:
        break
```

```python
f = "Data/train0.df.pkl.gz"
train = pd.read_pickle(f)
```

```python
t0 = time.time()
i = 1000
while True:
    f = "Data/train"+str(i)+".df.pkl.gz"
    try:
        x = pd.read_pickle(f)
        train = pd.concat([train, x])
        print("Finished ", f, i, i+1000, pretty_timedelta(since=t0))
        i += 1000
        if i == 26000:
            i = 27000 # skip bad file
    except FileNotFoundError:
        break
```

```python
def qtagged(p_quest, p_pos):
    pattern = re.compile("[^A-Za-z ]")
    qwords = pattern.sub("", p_quest).split()+["?"]
    qpos = p_pos.split()
    return list(zip(qwords,qpos))
```

```python
pos_dict = defaultdict(Counter)
word_dict = defaultdict(Counter)
for i in range(len(train)):
    qpair = train.iloc[[i]]
    for wpos in qtagged(qpair["question1"].item(), qpair["question1_pos"].item()):
        pos_dict[wpos[1]].update([wpos[0]])
        word_dict[wpos[0]].update([wpos[1]])
    for wpos in qtagged(qpair["question2"].item(), qpair["question2_pos"].item()):
        pos_dict[wpos[1]].update([wpos[0]])
        word_dict[wpos[0]].update([wpos[1]])        
```

```python
ic = ['Mumbai','Delhi','Bangalore','Hyderabad','Ahmedabad','Chennai','Kolkata','Surat','Pune','Jaipur']
tc = 0
for c in ic:
    cc = 0
    for k,v in enumerate(word_dict[c]):
        cc += word_dict[c][v]
    print(c,cc)
    tc += cc
print("India",tc)    
print(" ")

ic = ['York','Angeles','Chicago','Houston','Phoenix','Philadelphia','Antonio','Diego','Dallas','Jose']
tc = 0
for c in ic:
    cc = 0
    for k,v in enumerate(word_dict[c]):
        cc += word_dict[c][v]
    print(c,cc)    
    tc += cc
print("US", tc)
print(" ")

ic = ['Istanbul','Moscow','London','Budapest','Berlin','Madrid','Vienna','Rome','Paris','Barcelona']
tc = 0
for c in ic:
    cc = 0
    for k,v in enumerate(word_dict[c]):
        cc += word_dict[c][v]
    print(c,cc)    
    tc += cc
print("Europe", tc)
print(" ")

```
